```python
import pandas as pd
import nltk

# Download English stopwords
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))

# Define the remove_stop function
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]

# Read the CSV file into a DataFrame with the specified encoding
file_path = "C:/Users/geeth/OneDrive/Desktop/Special_Topics_DataScience/Statistics.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Apply the remove_stop function to the 'athlete' column and create a new column 'cleantokens'
df['cleantokens'] = df['athlete'].apply(lambda x: remove_stop(str(x).split()))

# Display the DataFrame with the cleaned tokens
print(df[['athlete', 'cleantokens']].tail())

```

### pipeline

```python
import pandas as pd
import nltk
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download English stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Define a custom transformer for tokenization
class Tokenizer(TransformerMixin):
    def transform(self, X, **transform_params):
        return [word_tokenize(text) for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

# Define a custom transformer for removing stopwords
class StopWordRemover(TransformerMixin):
    def transform(self, X, **transform_params):
        return [[word for word in text if word.lower() not in stop_words] for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

# Read the CSV file into a DataFrame
file_path = "C:/Users/geeth/OneDrive/Desktop/Special_Topics_DataScience/Statistics.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Define the pipeline
pipeline = Pipeline([
    ('tokenize', Tokenizer()),
    ('remove_stopwords', StopWordRemover())
])

# Apply the pipeline to the 'athlete' column
cleaned_text = pipeline.fit_transform(df['athlete'])

# Display the cleaned text
print(cleaned_text)
```

### Run the pipeline

```python
import pandas as pd
import nltk
from sklearn.pipeline import Pipeline
from sklearn.base import TransformerMixin
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download English stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Define a custom transformer for tokenization
class Tokenizer(TransformerMixin):
    def transform(self, X, **transform_params):
        return [word_tokenize(text) for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

# Define a custom transformer for removing stopwords
class StopWordRemover(TransformerMixin):
    def transform(self, X, **transform_params):
        return [[word for word in text if word.lower() not in stop_words] for text in X]

    def fit(self, X, y=None, **fit_params):
        return self

# Read the CSV file into a DataFrame
file_path = "C:/Users/geeth/OneDrive/Desktop/Special_Topics_DataScience/Statistics.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Define the pipeline
pipeline = Pipeline([
    ('tokenize', Tokenizer()),
    ('remove_stopwords', StopWordRemover())
])

# Apply the pipeline to the 'athlete' column
cleaned_text = pipeline.fit_transform(df['athlete'])

# Display the cleaned text
print(cleaned_text)
```

## Word Cloud

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
file_path = "C:/Users/geeth/OneDrive/Desktop/Special_Topics_DataScience/Statistics.csv"
# Read the CSV file into a DataFrame with a different encoding
df = pd.read_csv(file_path, encoding='ISO-8859-1')


# Concatenate text values from all relevant columns
text = ' '.join(df.astype(str).values.flatten())

# Define stopwords to exclude common words
stopwords = ["the", "and", "of", "in", "to", "a", "is", "it"]

# Create WordCloud with adjusted parameters
plt.figure(figsize=(14, 7))
wc = WordCloud(max_words=100, stopwords=stopwords, background_color='white',
               contour_width=0, contour_color='black', colormap='viridis')
wc.generate(text)

# Display the WordCloud
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

## word cloud in circular shape

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd

# Concatenate text values from all relevant columns
text = ' '.join(df.astype(str).values.flatten())

# Circle mask
x, y = np.ogrid[:300, :300]
mask = (x - 150) ** 2 + (y - 150) ** 2 > 130 ** 2
mask = 255 * mask.astype(int)

# Create WordCloud with a circle mask
wc = WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)

plt.figure(figsize=(8, 8))
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
from io import BytesIO
import requests
import pandas as pd

file_path = "C:/Users/geeth/OneDrive/Desktop/Special_Topics_DataScience/Statistics.csv"
# Read the CSV file into a DataFrame with a different encoding
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Concatenate text values from all relevant columns
text = ' '.join(df.astype(str).values.flatten())

# From URL
response = requests.get("https://365psd.com/images/istock/previews/1004/100424327-vector-cloud-icon.jpg")
mask = np.array(Image.open(BytesIO(response.content)))

# Colors
image_colors = ImageColorGenerator(mask)

wc = WordCloud(background_color="white", repeat=True, mask=mask)
wc.generate(text)

plt.figure(figsize=(8, 8))
plt.axis("off")
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")

plt.show()
```
